```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from scipy import stats
```

```python
#reading the csv files using pandas
student_mat = pd.read_csv('student-mat.csv')
student_por = pd.read_csv('student-por.csv')
```

```python
#The merged Data set reduced the number of rows and added a lot of excess info
student_data = pd.merge(student_mat,student_por, how = 'inner', on=[ "school","sex","age","address","famsize","Pstatus","Medu","Fedu","Mjob","Fjob","reason","nursery","internet"])
student_data.head()
```

```python
student_math = student_mat.rename(columns= {'school':'School',
                                           'famsize':'Family Size',
                                           'Pstatus': 'Parent Cohabitation Status',
                                           'Medu':"Mother's Education",
                                           'Fedu': "Father's Education",
                                           'Mjob': "Mother's Job",
                                           'Fjob': "Father's Job",
                                           'Dalc': 'Workday Alcohol Consumption',
                                           'Walc': 'Weekend Alcohol Consumption'})
student_math.head()
```

```python
#week consumption more intersting than weekday,expected for young kids
student_math['Workday Alcohol Consumption'].value_counts()
student_math['Weekend Alcohol Consumption'].value_counts()
```

```python
#might be better to define the columns earlier
math_age = student_math.iloc[:,2]

math_slope, math_int, math_r, math_p, math_std_err = stats.linregress(student_math['age'], student_math['Weekend Alcohol Consumption'])
math_fit = math_slope * math_age + math_int
```

```python
math_fit
```

```python
#only one 22 and 21 year old
math_age.value_counts()
```

```python
plt.xlim(min(math_age), max(math_age))
plt.plot(math_age,student_math['Weekend Alcohol Consumption'] , linewidth=1, marker="o")
plt.plot(math_age, math_fit, "b--", linewidth=1)
plt.ylabel("Drinking")
```
